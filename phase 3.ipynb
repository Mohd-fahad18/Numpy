```python
import numpy as np
import matplotlib.pyplot as plt
```

## Advance operation with business example

```python
# Data structure: [resturant_id , 2021 , 2022 , 2023 , 2024 ]
sales_data = np.array([
    [1 , 150000 , 180000 , 220000 , 250000], #paradise biryani
    [2 , 120000 , 140000 , 160000 , 190000], #chiken bite
    [3 , 200000 , 230000 , 260000 , 300000], #pizza hub
    [4 , 180000 , 210000 , 240000 , 320000], #burger point
    [5 , 160000 , 185000 , 205000 , 230000], #chai point
])

print("<=== Zomato sales analysys ===>")
print("\nsales data shape",sales_data.shape)

#selection of 2 rows
print("\n---sample data for first 2 restro data---\n",sales_data[0:2])

#section of 3 row except 1 elemnt of each
print("\n---sample data for 4 restro data---\n",sales_data[:4,1:])#first:for row 2 :for colm
```

```python
# TOTAL SALE PER YEAR 
print("sum of data column wise",np.sum(sales_data[:,1:],axis=0)) # 0 for column

# minimum sale per restro
min_sale = np.min(sales_data[:,1:],axis=1)
print("per restro minimum sale-->\n",min_sale)

# maximus sale per year
max_sale = np.max(sales_data[:,1:],axis=0)
print("max sale year wise \n",max_sale)

#AVERAGE SALE PER RESTRO
avg_sale = np.mean(sales_data[:,1:],axis=1)
print("avg sale per restro :\n",avg_sale)
```

```python
cumsum = np.cumsum(sales_data[:,1:],axis=1)#sales year wise and add last year
print(cumsum)

plt.Figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulative sale across all restro")
plt.xlabel("Year")
plt.ylabel("sales")
plt.grid(True)
plt.show
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector addition: ",vector1 +vector2)
print("Vector multiplication: ",vector1 * vector2)
print("\nDot product",np.dot(vector1,vector2))

angle = np.arccos(np.dot(vector1,vector2) / (np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print(angle) #linag use to find magnitude
```
